# Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure()
x=['Suzuki','Honda','BMW','Toyota']


r=np.arange(len(x))
width=0.25
height=3
yfemale=[30,50,40,20]
zmale=[40,20,60,70]
plt.bar(r-0.2,yfemale,0.4,color='b',label='female')
plt.bar(r+0.2,zmale,0.4, color='r',label='male')
plt.xticks(r,x)
plt.xlabel('Cars')
plt.ylabel('Age')
plt.title('Male vs Female buyers')
plt.legend()
plt.show()
```
